```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Given data
mean = 340
std_dev = 238.0602

# Generate x values
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)

# Calculate the corresponding y values for the normal distribution
y = norm.pdf(x, mean, std_dev)

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y, label='Bell Curve', color='blue')
plt.title('Bell Curve of Total Boys')
plt.xlabel('Total Boys')
plt.ylabel('Probability Density')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Function to plot bell curve
def plot_bell_curve(mean, std_dev, label, color):
    # Generate x values
    x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
    
    # Calculate the corresponding y values for the normal distribution
    y = norm.pdf(x, mean, std_dev)
    
    # Plotting the bell curve
    sns.lineplot(x=x, y=y, label=label, color=color)

# Given means and standard deviations for different groups
data = {
    'Total Boys': {
        'mean': 340,
        'std_dev': 238.0602,
        'color': 'blue'
    },
    'Total Girls': {
        'mean': 184.5,
        'std_dev': 134.1479,
        'color': 'orange'
    },
    'NA Boys': {
        'mean': 190.25,
        'std_dev': 152.1981,
        'color': 'green'
    },
    'NA Girls': {
        'mean': 109.75,
        'std_dev': 83.3921,
        'color': 'red'
    }
}

# Set up the plot
plt.figure(figsize=(20, 10))

# Plot each bell curve
for label, params in data.items():
    plot_bell_curve(params['mean'], params['std_dev'], label, params['color'])

# Add vertical lines for means
for label, params in data.items():
    plt.axvline(params['mean'], linestyle='--', color=params['color'], alpha=0.5)

# Set plot titles and labels
plt.title('Bell Curves for Different Groups')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Function to plot bell curve
def plot_bell_curve(mean, std_dev, label, color):
    # Generate x values
    x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
    
    # Calculate the corresponding y values for the normal distribution
    y = norm.pdf(x, mean, std_dev)
    
    # Plotting the bell curve
    sns.lineplot(x=x, y=y, label=label, color=color)
    
    # Highlight the maximum point
    max_index = np.argmax(y)
    plt.scatter(x[max_index], y[max_index], color='red', s=100, zorder=5)  # Highlight maximum point
    
    # Shade the area under the curve around the peak
    plt.fill_between(x, y, where=(x >= (mean - std_dev) * 0.75) & (x <= (mean + std_dev) * 0.75), color=color, alpha=0.3)

# Given means and standard deviations for different groups
data = {
    'Total Boys': {
        'mean': 340,
        'std_dev': 238.0602,
        'color': 'blue'
    },
    'Total Girls': {
        'mean': 184.5,
        'std_dev': 134.1479,
        'color': 'orange'
    },
    'NA Boys': {
        'mean': 190.25,
        'std_dev': 152.1981,
        'color': 'green'
    },
    'NA Girls': {
        'mean': 109.75,
        'std_dev': 83.3921,
        'color': 'red'
    }
}

# Set up the plot
plt.figure(figsize=(12, 8))

# Plot each bell curve
for label, params in data.items():
    plot_bell_curve(params['mean'], params['std_dev'], label, params['color'])

# Add vertical lines for means
for label, params in data.items():
    plt.axvline(params['mean'], linestyle='--', color=params['color'], alpha=0.5)

# Set plot titles and labels
plt.title('Bell Curves for Different Groups with Shaded Peaks')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()

```
